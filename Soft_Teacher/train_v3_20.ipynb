```python
import json
import os
import shutil

# Đường dẫn đến file JSON
json_path = "/home/coder/trong/KLTN_SEMI/code/Soft_Teacher/data_mmdetection/coco/semi_anns/instances_train2017.3@40.0.json"

# Thư mục ảnh gốc và thư mục đích
src_folder = "/home/coder/trong/KLTN_SEMI/code/Soft_Teacher/data_mmdetection/coco/train"
dst_folder = "/home/coder/trong/KLTN_SEMI/code/Soft_Teacher/data_mmdetection/coco/labeled_data1"

# Tạo thư mục đích nếu chưa có
os.makedirs(dst_folder, exist_ok=True)

# Đọc file JSON
with open(json_path, 'r') as f:
    data = json.load(f)

# Lấy danh sách file_name và di chuyển file
for image_info in data['images']:
    file_name = image_info['file_name']
    src_file = os.path.join(src_folder, file_name)
    dst_file = os.path.join(dst_folder, file_name)

    # Kiểm tra file tồn tại rồi mới di chuyển
    if os.path.exists(src_file):
        shutil.copy(src_file, dst_file)
        print(f"Đã di chuyển: {file_name}")
    else:
        print(f"Không tìm thấy file: {file_name}")

print(f"Hoàn thành việc di chuyển file.{dst_folder}")
```
